```python
# funcion llamada process_text() que recibe una cadena de texto y una lista de palabras a reemplazar

def process_text(text="Este es un texto de ejemplo.", palabras_a_reemplazar=None):
    if palabras_a_reemplazar is None:
      palabras_a_reemplazar = ["es", "texto"]

  # Convertir a minúsculas y eliminar espacios en blanco
    text = text.lower().strip()
  # Contador de palabras reemplazadas
    count_replaced = 0

    for palabra in palabras_a_reemplazar:
      if palabra in text:
            count_replaced += text.count(palabra)
            text = text.replace(palabra, '*' * len(palabra))
    return text, count_replaced

texto_original = " El texto a reemplazar es este "
palabras_de_reemplazo= ["texto", "es", "reemplazadas"]

nuevo_texto, cantidad_reemplazos = process_text(texto_original, palabras_de_reemplazo)
print(f"Texto procesado: '{nuevo_texto}'")
print(f"Cantidad de palabras reemplazadas: {cantidad_reemplazos}")
```

```python
#programa que realiza un análisis de números primos
#funcion que  determina si un número es primo
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

def prime_analysis():
  num = int(input("Ingrese un número entero positivo: "))
  if is_prime(num):
    print(f"{num} es un número primo.")
    print()
  else:
    print(f"{num} no es un número primo.")
    print()

  limite = int(input("Ingrese el límite para la búsqueda de primos: "))
  print("Los números primos menores que", limite, "son:")
  primes = []
  for i in range(2, limite):
    if is_prime(i):
      primes.append(i)
      print(i, end=" ")
  print()

  check_palindrome(primes)
  for prime in primes:
    categorize_prime(prime)

#funcion que determina si es palindromo
def check_palindrome(primes):
  print("Los números primos palíndromos son:")
  for num in primes:
    if str(num) == str(num)[::-1]:
      print(num, end=" ")
  print()

#funcion que clasifica el tipo de primo que es según su tamaño
def categorize_prime(prime):
  if prime < 10:
    print(f"{prime} es un número primo pequeño.")
    print()
  elif prime >= 10 and prime <= 99:
    print(f"{prime} es un número primo mediano.")
    print()
  else:
    print(f"{prime} es un número primo grande.")
    print()

if __name__ == "__main__":
  prime_analysis()
```

```python
def suma_total(numeros):
    return sum(numeros)

def suma_positivos_negativos(numeros):
    suma_positivos = sum(n for n in numeros if n > 0)
    suma_negativos = sum(n for n in numeros if n < 0)
    return suma_positivos, suma_negativos

def max_min(numeros):
    return max(numeros), min(numeros)

def son_unicos(numeros):
    return len(numeros) == len(set(numeros))

def mas_positivos_o_negativos(numeros):
    positivos = sum(1 for n in numeros if n > 0)
    negativos = sum(1 for n in numeros if n < 0)
    return "positivos" if positivos > negativos else "negativos" if negativos > positivos else "iguales"

def mostrar_menu():
    print("\nMenú:")
    print("1. Calcular la suma de todos los números")
    print("2. Calcular la suma de números positivos y negativos por separado")
    print("3. Encontrar el número máximo y mínimo de la lista")
    print("4. Comprobar si todos los números de la lista son únicos")
    print("5. Verificar si hay más números positivos o negativos")
    print("6. Salir")

def main():
    numeros = []

    # Leer los números del usuario
    while True:
        entrada = input("Introduce un número entero (o 'fin' para terminar): ")
        if entrada.lower() == 'fin':
            break
        try:
            numeros.append(int(entrada))
        except ValueError:
            print("Por favor, introduce un número entero válido.")

    while True:
        mostrar_menu()
        opcion = input("Selecciona una opción: ")

        if opcion == '1':
            print("La suma total de los números es:", suma_total(numeros))
        elif opcion == '2':
            suma_positivos, suma_negativos = suma_positivos_negativos(numeros)
            print("Suma de números positivos:", suma_positivos)
            print("Suma de números negativos:", suma_negativos)
        elif opcion == '3':
            maximo, minimo = max_min(numeros)
            print("Número máximo:", maximo)
            print("Número mínimo:", minimo)
        elif opcion == '4':
            if son_unicos(numeros):
                print("Todos los números son únicos.")
            else:
                print("Hay números duplicados en la lista.")
        elif opcion == '5':
            resultado = mas_positivos_o_negativos(numeros)
            print("La lista tiene más números", resultado)
        elif opcion == '6':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, intenta de nuevo.")

if __name__ == "__main__":
    main()
```

```python
# Variable global para almacenar las tareas
tareas = []

def add_task(task):
    #Agrega una nueva tarea a la lista global de tareas
    tareas.append(task)
    print(f"Tarea '{task}' agregada.")

def remove_task(task):
    #Elimina una tarea de la lista global de tareas, si existe
    try:
        tareas.remove(task)
        print(f"Tarea '{task}' eliminada.")
    except ValueError:
        print(f"Tarea '{task}' no encontrada.")

def list_tasks(tareas):
    #Enumera todas las tareas almacenadas en la lista global de tareas
    if not tareas:
        print("No hay tareas pendientes.")
    else:
        print("Tareas pendientes:")
        for i, tarea in enumerate(tareas, start=1):
            print(f"{i}. {tarea}")

def task_manager():
    #Administra las operaciones de agregar, modificar y enumerar tareas
    def modify_task():
        #Modifica la primera tarea en la lista de tareas
        if tareas:
            print("Modificando la primera tarea...")
            print(f"Tarea original: '{tareas[0]}'")
            nueva_tarea = input("Ingrese la nueva tarea: ")
            tareas[0] = nueva_tarea
            print(f"Tarea modificada a: '{nueva_tarea}'")
        else:
            print("No hay tareas para modificar.")

    while True:
        print("\n--- Gestor de Tareas ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Listar tareas")
        print("4. Modificar primera tarea")
        print("5. Salir")

        opcion = input("Seleccione una opción (1-5):")

        if opcion == "1":
            tarea = input("Ingrese la tarea a agregar: ")
        if opcion == "2":
            tarea = input("Ingrese la tarea a eliminar: ")
        if opcion == "3":
            list_tasks()
        if opcion == "4":
            modify_task()
        if opcion == "5":
            print("¡Hasta luego!")
            break

def main():
    """Función principal que inicia el programa."""
    task_manager()

if __name__ == "__main__":
    main()
```

```python
def imprimir_cadenas_por_letra(cadenas, letra):
    print(f"Cadenas que comienzan con la letra '{letra}':")
    encontrado = False
    for cadena in cadenas:
        if cadena.startswith(letra):
            print(cadena)
            encontrado = True
    if not encontrado:
        print(f"No se encontraron cadenas que comiencen con la letra '{letra}'.")

def contar_cadenas(cadenas, cadena_buscada):
    contador = cadenas.count(cadena_buscada)
    print(f"La cadena '{cadena_buscada}' aparece {contador} veces en la lista.")

def encontrar_cadenas_largas(cadenas):
    if not cadenas:
        print("La lista está vacía.")
        return

    cadena_mas_larga = max(cadenas, key=len)
    cadena_mas_corta = min(cadenas, key=len)
    print(f"La cadena más larga es: '{cadena_mas_larga}'")
    print(f"La cadena más corta es: '{cadena_mas_corta}'")

def verificar_objetos(cadenas):
    cadena1 = input("Ingrese la primera cadena a comparar: ")
    cadena2 = input("Ingrese la segunda cadena a comparar: ")

    if cadena1 in cadenas and cadena2 in cadenas:
        if cadena1 is cadena2:
            print("Ambas cadenas son el mismo objeto.")
        else:
            print("Las cadenas son diferentes objetos.")
    else:
        print("Una o ambas cadenas no están en la lista.")

def verificar_longitud(cadenas):
    for cadena in cadenas:
        if len(cadena) > 10:
            print(f"Encontrada una cadena con longitud mayor a 10: '{cadena}'")
            break
    else:
        print("Ninguna cadena tiene longitud mayor que 10.")

def salir():
    print("Saliendo del programa...")
    return True

def menu():
    cadenas = []

    # Leer la lista de cadenas desde teclado
    cantidad = int(input("¿Cuántas cadenas desea ingresar? "))
    for _ in range(cantidad):
        cadena = input("Ingrese una cadena: ")
        cadenas.append(cadena)

    while True:
        print("\nMenú de opciones:")
        print("1. Imprimir cadenas que comienzan con una letra específica")
        print("2. Contar cuántas veces aparece una cadena específica")
        print("3. Encontrar la cadena más larga y la más corta")
        print("4. Verificar si dos cadenas son el mismo objeto")
        print("5. Verificar longitud de cadenas")
        print("6. Salir")

        opcion = input("Seleccione una opción (1-6): ")

        if opcion == '1':
            letra = input("Ingrese la letra: ")
            imprimir_cadenas_por_letra(cadenas, letra)
        elif opcion == '2':
            cadena_buscada = input("Ingrese la cadena a contar: ")
            contar_cadenas(cadenas, cadena_buscada)
        elif opcion == '3':
            encontrar_cadenas_largas(cadenas)
        elif opcion == '4':
            verificar_objetos(cadenas)
        elif opcion == '5':
            verificar_longitud(cadenas)
        elif opcion == '6':
            if salir():
                break
        else:
            print("Opción no válida. Por favor, elija una opción del menú.")

if __name__ == "__main__":
    menu()
```

Explicación del Programa
Funciones:

imprimir_cadenas_por_letra(cadenas, letra): Imprime todas las cadenas que comienzan con una letra específica.

contar_cadenas(cadenas, cadena_buscada): Cuenta cuántas veces aparece una cadena específica en la lista.

encontrar_cadenas_largas(cadenas): Encuentra e imprime la cadena más larga y la más corta de la lista.

verificar_objetos(cadenas): Verifica si dos cadenas específicas de la lista son el mismo objeto usando el operador is.

verificar_longitud(cadenas): Usa un bucle for...else para verificar si alguna cadena tiene longitud mayor a 10.

salir(): Imprime un mensaje de salida y devuelve True.

Función menu(): Contiene el bucle principal del programa, donde se presentan las opciones al usuario y se llaman a las funciones correspondientes según la elección.

Entrada de Cadenas: Antes de mostrar el menú, el programa solicita al usuario que ingrese cuántas cadenas desea agregar y las lee desde el teclado.
